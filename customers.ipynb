```python
import pandas as pd
import time
import numpy as np
import seaborn as sns
```

```python
df_customers = pd.read_csv('10. Projekt Model uporabnikov z Markovsko verigo (1)/EMS_Customer_Satisfaction_By_Month.csv')
len(df_customers)
```

```python
df_customers.head(5)
```

```python

df_customers['Month-Year'] = df_customers['Month-Year'].str.replace(' 12:00:00 AM', '')
time_index = pd.date_range('10/01/2013', periods=len(df_customers),  freq='MS') 
time_index = pd.DatetimeIndex(time_index)
df_customers = df_customers.set_index(time_index)
df_customers = df_customers.drop(['Month-Year'], axis=1)
df_customers.iloc[np.r_[0:5,-5:0]].iloc[:,0]
df_customers.head(5)
```
